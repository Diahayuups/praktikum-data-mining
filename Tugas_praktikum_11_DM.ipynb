```python
# mengimport pandas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
# membaca dataset
mydata = pd.read_csv('/content/drive/MyDrive/praktikum_dm/dataset/temperature_2015.csv')
mydata
```

```python
# mengecek typedata
mydata.info()
```

```python
# mengubah type data kolom index menjadi object
mydata['index'] = mydata['index'].astype(object)
mydata.info()
```

```python
# prompt: # mengubah tipe data kolom JAN FEB MAR APR MAY JUN JUL AUG SEP OCT NOV DEC ANN menjadi numeric float

# List of columns to convert to numeric
cols_to_convert = ['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC', 'ANN']

# Iterate through the columns and convert them to numeric, handling errors
for col in cols_to_convert:
    mydata[col] = pd.to_numeric(mydata[col], errors='coerce')

# Check the data types after conversion
mydata.info()
```

```python
# mencari hasil dari statistika deskriptif kolom float
mydata.describe()
```

```python
# mengecek missing value
missing_value = mydata.isnull().sum()
missing_value
```

```python
#meghitung ratio missing value pada kolom
rasio = missing_value.div(len(mydata))
rasio
```

```python
# mengisi missing value kolom int (JAN FEB MAR APR MAY JUN  JUL AUG SEP OCT NOV DEC ANN) dengan median
mydata['JAN'].fillna(mydata['JAN'].median(), inplace=True)
mydata['FEB'].fillna(mydata['FEB'].median(), inplace=True)
mydata['MAR'].fillna(mydata['MAR'].median(), inplace=True)
mydata['APR'].fillna(mydata['APR'].median(), inplace=True)
mydata['MAY'].fillna(mydata['MAY'].median(), inplace=True)
mydata['JUN'].fillna(mydata['JUN'].median(), inplace=True)
mydata['JUL'].fillna(mydata['JUL'].median(), inplace=True)
mydata['AUG'].fillna(mydata['AUG'].median(), inplace=True)
mydata['SEP'].fillna(mydata['SEP'].median(), inplace=True)
mydata['OCT'].fillna(mydata['OCT'].median(), inplace=True)
mydata['NOV'].fillna(mydata['NOV'].median(), inplace=True)
mydata['DEC'].fillna(mydata['DEC'].median(), inplace=True)
mydata['ANN'].fillna(mydata['ANN'].median(), inplace=True)
```

```python
# mengecek missing value lagi
missing_value = mydata.isnull().sum()
missing_value
```

```python
# Regresi
x = mydata[["JAN", "FEB", "MAR"]].to_numpy()  # Gunakan bulan sebagai fitur
y = mydata["ANN"].to_numpy()  # Rata-rata tahunan sebagai target

x = sm.add_constant(x)
model_prediksi = sm.OLS(y, x).fit()
print(model_prediksi.summary())
```
