```python
# mengimport dan menampilkan dataset

import pandas as pd
import numpy as np

mydata = pd.read_csv("BL-Flickr-Images-Book.csv")
mydata
```

```python
# melihat datatype dan spesfikasi data
mydata.info()
```

```python
# mengecek missing values data
mydata.isnull().sum()
```

```python
# menghapus data (kolom) tertentu yang tidak diperlukan karena keseluruhan data memiliki daa kosong

columns_to_drop = ['Edition Statement', 'Corporate Author', 'Corporate Contributors', 'Former owner', 'Engraver']

mydata.drop(columns=columns_to_drop, inplace=True)
mydata
```

```python
# menjadikan indentifier sebagai index agar memudahkan mengambil sebuah data
mydata.set_index('Identifier', inplace=True)
mydata
```

```python
# merubah nama kolom
mydata.rename(columns={
    'Place of Publication': 'Tempat_Penerbitan',
    'Date of Publication': 'Tahun_Penerbitan',
    'Publisher': 'Penerbit',
    'Title': 'Judul',
    'Author': 'Penulis',
    'Contributors': 'Kontributor',
    'Issuance type': 'Jenis_Penerbitan',
    'Flickr URL': 'URL_Flickr',
    'Shelfmarks': 'Penanda_Rak'
}, inplace=True)

# menampilkan dataframe setelah di rename
mydata
```
